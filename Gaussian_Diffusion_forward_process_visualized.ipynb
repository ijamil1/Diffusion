```python
import numpy as np
from matplotlib.pyplot import hist
```

```python
from scipy.stats import norm, rayleigh
```

```python
values = hist(rayleigh.rvs(loc = 0, scale = 80, size = 100000) - 80 * np.sqrt(np.pi/2), bins=200, density=True)
```

```python
np.mean(rayleigh.rvs(loc = 0, scale = 80, size = 100000) - 80 * np.sqrt(np.pi/2))
```

```python
std_dev = 0.001
T = 10 ** 10
num_samples = 10 ** 6

# Vectorized sampling
rayleigh_samples = rayleigh.rvs(loc=0, scale=80, size=num_samples) - 80 * np.sqrt(np.pi/2)
samples = norm.rvs(loc=rayleigh_samples, scale=std_dev * np.sqrt(T))

print(np.mean(samples))
print(np.std(samples))

```

```python
hist(samples, bins = 100, density = True)
```

```python
hist(np.random.uniform(-1, 1, 100000), density=True)
```

```python
std_dev = 0.001
T = 10 ** 10
num_samples = 10 ** 6

# Vectorized sampling
uniform_samples = np.random.uniform(-1, 1, num_samples)
samples = norm.rvs(loc=uniform_samples, scale=std_dev * np.sqrt(T))

print(np.mean(samples))
print(np.std(samples))

```

```python
hist(samples, bins = 100, density = True)
```

```python
mu1 = 50
mu2 = 100

std_dev1 = 5
std_dev2 = 10

size = 10000000

samples1 = norm.rvs(loc=mu1, scale=std_dev1, size=size)
samples2 = norm.rvs(loc=mu2, scale=std_dev2, size=size)


selector = np.random.rand(size) < 0.5  # True  use samples1, False  use samples2

# Vectorized selection
mixture_samples = np.where(selector, samples1, samples2)

print(np.mean(mixture_samples))
print(np.std(mixture_samples))

values = hist(mixture_samples, bins=200, density=True)


std_dev = 0.001
T = 10 ** 10
num_samples = 10 ** 6

# Vectorized sampling
samples = norm.rvs(loc=mixture_samples, scale=std_dev * np.sqrt(T))

print(np.mean(samples))
print(np.std(samples))

```

```python
hist(samples, bins = 200, density=True)
```
